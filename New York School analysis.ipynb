```python
#importing pandas
import pandas as pd

# reading the data
schools = pd.read_csv("schools.csv")

# reviewing data
schools.info()
schools.head()
```

```python
# NYC school with the best math result
best_math_schools = schools[schools["average_math"]>= 640]

sorted_schools = best_math_schools.sort_values(["school_name", "average_math"], ascending=[False, True])

best_math_schools = sorted_schools[["school_name", "average_math"]]
print(best_math_schools)
```

```python
# Top 10 performing schools based on combines SAT score.

#new column total_SAT
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# sorting the top 10 schools
top_10_schools = schools.sort_values("total_SAT", ascending = False)[["school_name", "total_SAT"]].head(10)
print(top_10_schools)
```
